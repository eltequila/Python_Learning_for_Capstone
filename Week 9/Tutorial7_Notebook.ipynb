Today, we are going to plot a pie chart.

Our first step is to import out Python Libraries. 

```python
from pathlib import Path
import csv
import matplotlib.pyplot as plt
```

Our second step is to load our data. 

```python
csv_path = Path(r'/Users/macbookairfromboeing/Downloads/Penn/Courses/2025 Fall/ENVS 5726/Week 05 - Summarizations/Week05TutorialData/EMDAT_Haiti_Disaster_Summary.csv')

emdat_summary_table = []
with open (csv_path, 'r', encoding = 'utf-8') as csv_file:
    reader = csv.reader(csv_file)
    headers = next(reader)
    for row in reader:
        emdat_summary_table.append(row)

category_column = [row[headers.index('Disaster Type')] for row in emdat_summary_table]
value_column = [row[headers.index('Total Death')] for row in emdat_summary_table]
```

```python
plt.pie(value_column, labels = category_column)
plt.title('Total Death by Natural Disaster Type')
plt.legend(bbox_to_anchor=(1.2,0.6), loc="upper left")
plt.show()
```
